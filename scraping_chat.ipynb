```python
import re
import requests
from bs4 import BeautifulSoup

# Define the URL of the webpage you want to scrape
url = "https://bester-eats.com/area/PRO64/ARE829/LCAT29/CAT1123/"

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    mysoup = BeautifulSoup(response.text, 'html.parser')

    # map_start = "https://bester-eats.com/restaurant/91331/map/"
    # map_start = "https://www.google.com/maps/place/"

    name_elements = mysoup.find_all("a", {"class": "restaurant-card-header"})
    price_elements = mysoup.find_all("span", {"class":"cost-list__item cost-item cost-item--undefined"})
    address_elements = mysoup.find_all("div", {"class": "restaurant-card__address"})
    tell_elements = mysoup.find_all("div", {"class": "restaurant-card__phone"})
    # map_elements = mysoup.find_all("a", href=True)
    reviews_elements = mysoup.find_all("p", {"class": "review-item__content"})

    my_corpus = []

    for i in range(len(name_elements)):
        name = name_elements[i].text.strip()
        price = price_elements[i].text.strip()
        address = address_elements[i].text.strip()


        # Clean up phone numbers by removing non-digit characters
        cleaned_number = re.sub(r'\D', '', tell_elements[i].text.strip())
        if cleaned_number.startswith('66'):
          cleaned_number = cleaned_number[2:]
        tell = '0' + cleaned_number

        # map_location = map_elements[i]['href']
        reviews = reviews_elements[i].text.strip()

        # Check if the map location starts with the specified prefix
        # if map_location.startswith(map_start):
        #     maps = map_location
        # else:
        #     maps = "Google Maps link not found"

        my_corpus.append((name, price, address, tell, reviews))

    # Print the result
    for item in my_corpus:
        print("\nRestaurant name:", item[0])
        print("Price:", item[1])
        print("Address:", item[2])
        print("Tell:", item[3])
        print("Reviews:", item[4])

else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
import numpy as np

def get_similarity(question, corpus):
    #Compute embedding for both lists
    # question = "".join(question.split())
    entity_corpus = corpus
    question_vec = model.encode(question, convert_to_tensor=True, normalize_embeddings=True)
    corpus_vec = model.encode(entity_corpus, convert_to_tensor=True, normalize_embeddings=True)

    #Compute cosine-similarities
    cosine_scores = util.cos_sim(question_vec, corpus_vec)
    entity_np = np.array(cosine_scores)

    #Output the pairs with their score
    score = np.max(entity_np)
    # print(f"Max score = {score:.2f}")
    if score >= 0.6:
      match_entity = entity_corpus[np.argmax(entity_np)]
      # print("Output :", match_entity)
      return [match_entity, score]
    else:
      # print(f"‡∏Ç‡∏≠‡∏≠‡∏†‡∏±‡∏¢ ‡∏â‡∏±‡∏ô‡πÑ‡∏°‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡πÉ‡∏à‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏≤‡∏°‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì")
      return [f"‡∏Ç‡∏≠‡∏≠‡∏†‡∏±‡∏¢‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡πà‡∏∞ ‡∏â‡∏±‡∏ô‡πÑ‡∏°‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡πÉ‡∏à‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏≤‡∏°‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì", score]

def get_contact(contact_input):
  contact_input = ["yes", "no"]
  contact_result = get_similarity(contact_input, contact_input)
  return contact_result

def get_price(price_input):
  price_corpus = ["yes", "no"]
  price_result = get_similarity(price_input, price_corpus)
  return price_result

def get_type(type_input):
  type_corpus = ["‡∏Å‡πã‡∏ß‡∏¢‡πÄ‡∏ï‡∏µ‡πã‡∏¢‡∏ß", "‡∏ö‡∏∞‡∏´‡∏°‡∏µ‡πà", "‡πÄ‡∏¢‡πá‡∏ô‡∏ï‡∏≤‡πÇ‡∏ü"]
  type_result = get_similarity(type_input, type_corpus)
  return type_result

def get_search_terms(inputs):
  search_terms = []
  # type term
  search_type = get_type(inputs[0])
  search_terms.append(search_type)
  # price term
  search_type = get_price(inputs[1])
  search_terms.append(search_type)
  # contact term
  search_type = get_contact(inputs[2])
  search_terms.append(search_type)
  return search_terms

def get_answer(input1, input2, input3):
  # filter input with cosine similarity
  inputs = [input1, input2, input3]
  search_terms = get_search_terms(inputs)
  # print("Search Terms and Score :", search_terms, "\n")
  # match and recommend
  keywords = [s[0] for s in search_terms]
  mask = (df_filter["Type"] == keywords[0]) & (df_filter["Have Price details?"] == keywords[1]) & (df_filter["Have Contact?"] == keywords[2])
  df_mask = df_filter.copy()
  df_mask.loc[mask, "Recommend?"] = "yes"
  df_mask.loc[~mask, "Recommend?"] = "no"

  df_match = df_mask.loc[df_mask["Recommend?"] == "yes", :]
  df_drop = df_match.drop(["Type", "Have Contact?", "Have Price details?", "Recommend?"], axis=1)
  df_answer = df_drop.to_dict(orient="records")
  answer_str = ""
  for i in range(len(df_answer)):
    answer_str += f"{i+1}. "
    for key, value in df_answer[i].items():
      answer_str += (f"{key} : {value}" + "\n")
    answer_str += "\n"
  # return [answer_str, search_terms]
  return answer_str

# input1, input2, input3 = "‡∏£‡πâ‡∏≤‡∏ô‡πÄ‡∏´‡∏•‡πâ‡∏≤", "‡πÑ‡∏°‡πà", "‡πÉ‡∏ä"
# result = get_answer(input1, input2, input3)
# print("Search Terms and Score :", result[1])
# print()
# print(result[0])
```
